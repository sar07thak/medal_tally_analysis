```python
import pandas as pd 
import numpy as np 
```

```python
path_athelete = "../data/athlete_events.csv"
path_region = "../data/noc_regions.csv"

df = pd.read_csv(path_athelete)
region_df = pd.read_csv(path_region)
```

```python
print( df.columns )
print( region_df.columns)
```

```python
df = df.merge( region_df , on = 'NOC' , how='inner' )
df = df[df['Season'] == 'Summer']
df.head()
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates( inplace= True )
df.duplicated().sum()
```

```python
df = pd.concat([df , pd.get_dummies(df['Medal']) ] , axis=1)
df.head()
```

```python
# Step 1: Keep only medal winners
df_medal = df[df['Medal'].notnull()]

# Step 2: Group by Name + Sport + Region
medal_sport_df = df_medal.groupby(
    ['Name', 'Sport', 'region']
)[['Gold', 'Silver', 'Bronze']].sum().reset_index()

# Step 3: Add total medals
medal_sport_df['Total'] = (
    medal_sport_df['Gold'] +
    medal_sport_df['Silver'] +
    medal_sport_df['Bronze']
)

# Step 4: Sort
medal_sport_df = medal_sport_df.sort_values(by='Total', ascending=False)

table = medal_sport_df[['Name','Sport','region','Gold' ,'Silver','Bronze','Total']]
table.reset_index().drop('index',axis=1)



# Step 5: Optional → top 10
# top_players = medal_sport_df

# top_players.head(15)


# top_players[top_players['Sport'] == 'Gymnastics'].head(15)
```
