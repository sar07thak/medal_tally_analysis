```python
import pandas as pd 
data = {
    't_id' : [ 1,1,1,2,2,2,2] ,
    's_id' : [2,2,3,1,2,3,4] ,
    'd_id' : [3,4,3,1,1,1,1]
}

df = pd.DataFrame( data )

df = df.drop_duplicates(subset=['t_id','s_id'])
df = df.groupby('t_id').size().reset_index().rename(columns={-1: 'count'})
df
```

```python
import pandas as pd 

data = {
    "name": ["A", "B", "C", "D"],
    "marks": [85, 92, 78, 92]
}


df = pd.DataFrame( data ) 
df = df.sort_values(by="marks", ascending=False)
df['rank'] = df['marks'].rank( ascending=False , method="dense")
df
```
