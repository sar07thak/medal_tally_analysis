```python
df.groupby('NOC')[['Gold','Silver','Bronze']].sum() \
  .sort_values(by='Gold', ascending=False) \
  .reset_index()

# In this data set there is mistake count medal multiple times as they held in a 
# same year of same sport like a hockey team win a gold in a specific year so count as 1 but it count as = Number of players*Gold
```

```python
# This is the problem for india win a GOLD medal in hockey and it count 14 because hockey has 14 total num of player
df[ (df['NOC'] == 'IND')  &  (df['Medal'] == 'Gold' ) & ( df['Sport'] == 'Hockey') & ( df['Year'] == 1928 )].head()
```

```python
# Fix: Count team medals as 1 medal per event, not per player
# Drop duplicates based on the EVENT won (NOC + Year + Sport + Event + Medal)
# This ensures: Team sports (Hockey, Basketball) = 1 medal, Individual sports = 1 medal per athlete
df = df.drop_duplicates(subset=['NOC','Games','Year','Season','City','Sport','Event','Medal','region','notes','Bronze','Gold','Silver'])
df.head()
```

```python
# Verify: India's gold medals should now be much lower (team sports counted as 1)
# Before fix: 131 (counted each hockey player individually)
# After fix: Should count each unique event only once
df[ (df['NOC'] == 'IND')  &  (df['Medal'] == 'Gold' )].shape
```

```python
# Verify hockey example: India 1928 Gold should now be 1 medal, not 14 (one per player)
print('India 1928 Hockey Gold - rows after fix:')
print(df[ (df['NOC'] == 'IND') & (df['Medal'] == 'Gold') & (df['Sport'] == 'Hockey') & (df['Year'] == 1928) ].shape)
print('\nBefore fix this was 14 rows (one per player), now it should be 1 row (one team medal)')
df[ (df['NOC'] == 'IND') & (df['Medal'] == 'Gold') & (df['Sport'] == 'Hockey') & (df['Year'] == 1928) ].head()
```

```python
# Now this medal tally has correct data about the medal with specific nation.
# Team sports (like Hockey, Basketball) now count as 1 medal per event, not per player.
df.shape
```
